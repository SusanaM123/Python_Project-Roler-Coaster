```python
#Subseting
df = df[['coaster_name','Manufacturer', 'year_introduced', 'latitude', 'longitude',
    'Type_Main', 'opening_date_clean','speed_mph', 'height_value',
    'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df.dtypes
```

```python
#change data type
df['opening_date_clean'] = df['opening_date_clean'].apply(pd.to_datetime)
```

```python
df.dtypes
```

```python
#Rename Columns
df = df.rename(columns={'coaster_name':'Coaster_Name',
           'year_introduced':'Year_Introduced',
           'latitude':'Latitude',
           'longitude':'Longitude',
           'opening_date_clean':'Opening_Date',
           'speed_mph':'Speed_Mph',
           'height_value':'Height_ft',
           'Inversions_clean':'Inversions',
           'Gforce_clean':'Gforce'})
```

```python
df.head()
```

```python
#null values
df.isna().sum()
```

```python
#duplicates
df.duplicated().sum()
#opcion 2
df.loc[df.duplicated()]
```

```python
#double check for duplicates
df.duplicated(subset=['Coaster_Name','Latitude', 'Longitude','Opening_Date']).sum()
#
```

```python
df.loc[df.duplicated(subset=['Coaster_Name','Latitude', 'Longitude','Opening_Date'])]
```

```python
df = df.loc[~df.duplicated(subset=['Coaster_Name','Latitude', 'Longitude','Opening_Date'])]\
.reset_index().copy()
```

```python
df.shape
```

```python
# GRAPHS
df['Year_Introduced'].value_counts().head(10).plot(kind='bar',
                                                   title = 'Coasters Introduced per Year ',
                                                   xlabel= 'Year', ylabel='Count')
```

```python
df['Speed_Mph'].plot(kind='hist',bins=40,
                     title='Coaster Speed in mph',
                     xlabel='Speed')
```

```python
df['Speed_Mph'].plot(kind='kde',
                     title='Coaster Speed in mph',
                     xlabel='Speed')
```

```python
df = df.fillna(0)
```

```python
df['Height_ft'] = df['Height_ft'].astype('int')
```

```python
df = df.drop(836,axis = 0)
```

```python
# another SCOTTER PLOT OPTION
sns.scatterplot(data=df, x='Speed_Mph', y='Height_ft',hue='Year_Introduced')
plt.show()
```
